```python
# Function to calculate and accumulate evaluation metrics
def evaluate_model(y_true, y_pred):
    accuracy = accuracy_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred, pos_label='yes')
    recall = recall_score(y_true, y_pred, pos_label='yes')
    f1 = f1_score(y_true, y_pred, pos_label='yes')
    return accuracy, precision, recall, f1

# Function to plot confusion matrix
def plot_confusion_matrix(cm, model_name):
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['yes', 'no'], yticklabels=['yes', 'no'])
    plt.title(f'Confusion Matrix: {model_name}')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.show()

# Function to plot loss curves
def plot_loss_curve(train_loss, test_loss, model_name):
    plt.figure(figsize=(10, 6))
    plt.plot(train_loss, label='Train Loss', color='blue')
    plt.plot(test_loss, label='Test Loss', color='red')
    plt.title(f'Loss Curve for {model_name}')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
    plt.show()
```

## **Replacing NAN values & Precessing by Z scores**

```python
# Drop unnecessary columns
bc_tcga_data = bc_tcga.drop(columns=['Hybridization REF', 'tumor'])

# Separate the data into tumor and non-tumor groups
tumor_data = bc_tcga_data[bc_tcga['tumor'] == 'yes']
normal_data = bc_tcga_data[bc_tcga['tumor'] == 'no']

# Calculate mean values for each group
tumor_mean = tumor_data.mean()
normal_mean = normal_data.mean()

# Fill missing values with mean values
bc_tcga_data_tumor_filled = tumor_data.fillna(tumor_mean)
bc_tcga_data_normal_filled = normal_data.fillna(normal_mean)

# Concatenate the filled data back together
bc_tcga_data_filled = pd.concat([bc_tcga_data_tumor_filled, bc_tcga_data_normal_filled])

# Get the 'Hybridization REF' column
hybridization_ref = bc_tcga['Hybridization REF']

# Add 'Hybridization REF' and 'tumor' columns back
bc_tcga_data_filled = pd.concat([hybridization_ref, bc_tcga_data_filled, bc_tcga['tumor']], axis=1)

# Save the preprocessed data
bc_tcga_data_filled.to_csv('Preprocessed_BC-TCGA_Filled.csv', index=False)

# Apply Z-score normalization
data_values = bc_tcga_data_filled.iloc[:, 1:-1].values
scaler = StandardScaler()
normalized_data = scaler.fit_transform(data_values)

# Create a DataFrame with the normalized values
bc_tcga_normalized = pd.DataFrame(data=normalized_data, columns=bc_tcga_data_filled.columns[1:-1])

# Add 'Hybridization REF' and 'tumor' columns back
bc_tcga_normalized = pd.concat([hybridization_ref, bc_tcga_normalized, bc_tcga['tumor']], axis=1)

# Save the preprocessed data
bc_tcga_normalized.to_csv('Preprocessed_BC-TCGA_Normalized.csv', index=False)

```

```python
kf = StratifiedKFold(n_splits=10, shuffle=True, random_state=15)

# Store metrics
accuracies = []
train_losses = []
val_losses = []

# For each fold, calculate the losses
for train_index, val_index in kf.split(X_pca, y):
    X_train_cv, X_val_cv = X_pca[train_index], X_pca[val_index]
    y_train_cv, y_val_cv = y.iloc[train_index], y.iloc[val_index]

    # Fit the model
    svm_model = SVC(C=0.0001, probability=True)  # Set probability=True to enable predict_proba
    svm_model.fit(X_train_cv, y_train_cv)

    # Predict on validation set
    y_val_pred = svm_model.predict(X_val_cv)

    # Calculate accuracy
    accuracy = accuracy_score(y_val_cv, y_val_pred)
    accuracies.append(accuracy)

    # Log loss
    val_probas = svm_model.predict_proba(X_val_cv)
    val_loss = log_loss(y_val_cv, val_probas)
    val_losses.append(val_loss)

    # Train loss (using predictions on training set)
    train_probas = svm_model.predict_proba(X_train_cv)
    train_loss = log_loss(y_train_cv, train_probas)
    train_losses.append(train_loss)

# Calculate average accuracy and log loss
avg_accuracy = np.mean(accuracies)
avg_val_loss = np.mean(val_losses)
avg_train_loss = np.mean(train_losses)

print(f'Average Validation Accuracy: {avg_accuracy * 100:.2f}%')
print(f'Average Training Log Loss: {avg_train_loss:.4f}')
print(f'Average Validation Log Loss: {avg_val_loss:.4f}')

# Confusion Matrix and Classification Report for the last fold
cf_matrix = confusion_matrix(y_val_cv, y_val_pred)
print('\nConfusion Matrix:')
print(cf_matrix)
print('\nClassification Report:')
print(classification_report(y_val_cv, y_val_pred))

# Plotting the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cf_matrix, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()

# Plotting Loss Curves
epochs = range(1, len(val_losses) + 1)

plt.figure(figsize=(12, 6))
plt.plot(epochs, train_losses, label='Training Log Loss', marker='o')
plt.plot(epochs, val_losses, label='Validation Log Loss', marker='x')
plt.title('Training and Validation Log Loss')
plt.xlabel('Fold Number')
plt.ylabel('Log Loss')
plt.legend()
plt.show()
```

```python
from keras.models import Sequential
from keras.layers import Dense, GaussianNoise, Dropout
from keras.optimizers import Adam
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, log_loss
from sklearn.model_selection import StratifiedKFold
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Encode target variable
label_encoder = LabelEncoder()
bc_tcga_normalized['tumor'] = label_encoder.fit_transform(bc_tcga_normalized['tumor'])

# Define features and target variable
X = bc_tcga_normalized.drop(columns=['Hybridization REF', 'tumor'])
y = bc_tcga_normalized['tumor']

# Handle NaN values by dropping rows
X.dropna(inplace=True)
y = y[X.index]  # Align y with X after dropping rows

# Normalize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply SDAE (Stacked Denoising Autoencoder)
sdae = Sequential()
sdae.add(GaussianNoise(0.1, input_shape=(X_scaled.shape[1],)))  # Add Gaussian noise to input
sdae.add(Dense(200, activation='relu'))
sdae.add(GaussianNoise(0.1))
sdae.add(Dense(100, activation='relu'))
sdae.add(Dense(X_scaled.shape[1], activation='linear'))  # Reconstruct to original dimensions

sdae.compile(optimizer='adam', loss='mean_squared_error')

# Train SDAE
sdae.fit(X_scaled, X_scaled, epochs=50, batch_size=32, shuffle=True, verbose=0)

# Transform the input data using the trained SDAE
X_sdae = sdae.predict(X_scaled)

# Stratified K-Fold Cross-Validation
kf = StratifiedKFold(n_splits=5, shuffle=True, random_state=5)

# Store metrics
accuracies = []
train_losses = []
val_losses = []

# For each fold, calculate the losses and accuracy
for train_index, val_index in kf.split(X_sdae, y):
    X_train_cv, X_val_cv = X_sdae[train_index], X_sdae[val_index]
    y_train_cv, y_val_cv = y.iloc[train_index], y.iloc[val_index]

    # Build the ANN model
    model = Sequential()
    model.add(Dense(128, input_dim=X_train_cv.shape[1], activation='relu'))
    model.add(Dropout(0.1))
    model.add(Dense(64, input_dim=X_train_cv.shape[1], activation='relu'))
    model.add(Dropout(0.1))
    model.add(Dense(32, activation='relu'))
    model.add(Dropout(0.1))
    model.add(Dense(1, activation='sigmoid'))  # Binary classification

    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

    # Fit the model
    history = model.fit(X_train_cv, y_train_cv, epochs=50, batch_size=64,
                        validation_data=(X_val_cv, y_val_cv), verbose=0)

    # Predict on validation set
    y_val_pred = model.predict(X_val_cv)
    y_val_pred_classes = (y_val_pred > 0.5).astype(int)

    # Calculate accuracy
    accuracy = accuracy_score(y_val_cv, y_val_pred_classes)
    accuracies.append(accuracy)

    # Log loss
    val_loss = log_loss(y_val_cv, y_val_pred)
    val_losses.append(val_loss)

    # Train loss (using predictions on training set)
    train_probas = model.predict(X_train_cv)
    train_loss = log_loss(y_train_cv, train_probas)
    train_losses.append(train_loss)

# Calculate average accuracy and log loss
avg_accuracy = np.mean(accuracies)
avg_train_loss = np.mean(train_losses)
avg_val_loss = np.mean(val_losses)

print(f'Average Validation Accuracy: {avg_accuracy * 100:.2f}%')
print(f'Average Training Log Loss: {avg_train_loss:.4f}')
print(f'Average Validation Log Loss: {avg_val_loss:.4f}')

# Confusion Matrix and Classification Report for the last fold
cf_matrix = confusion_matrix(y_val_cv, y_val_pred_classes)
print('\nConfusion Matrix:')
print(cf_matrix)
print('\nClassification Report:')
print(classification_report(y_val_cv, y_val_pred_classes))

# Plotting the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cf_matrix, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()

# Plotting Loss Curves
epochs = range(1, len(history.history['loss']) + 1)

plt.figure(figsize=(12, 6))
plt.plot(epochs, history.history['loss'], label='Training Log Loss', marker='o')
plt.plot(epochs, history.history['val_loss'], label='Validation Log Loss', marker='x')
plt.title('Training and Validation Log Loss')
plt.xlabel('Epochs')
plt.ylabel('Log Loss')
plt.legend()
plt.show()
```

```python
# Define constants
dim = 1000
iterations = 10

# Initialize dictionaries to store average metrics
avg_metrics = {
    'SVM': {'accuracy': [], 'precision': [], 'recall': [], 'f1': []},
    'SVM-RBF': {'accuracy': [], 'precision': [], 'recall': [], 'f1': []},
    'ANN': {'accuracy': [], 'precision': [], 'recall': [], 'f1': []}
}

def train_sdae(X_train_scaled, X_test_scaled):
    sdae = Sequential()
    sdae.add(GaussianNoise(0.1, input_shape=(X_train_scaled.shape[1],)))  # Add Gaussian noise
    # Define the SDAE architecture
    layers = [800, 400, 200, 100, 50, 25, 50, 100, 200, 400, 800]
    for units in layers:
        sdae.add(Dense(units=units, activation='relu'))
        sdae.add(GaussianNoise(0.1))
    sdae.add(Dense(units=X_train_scaled.shape[1], activation='linear'))

    sdae.compile(optimizer='adam', loss='mean_squared_error')
    sdae.fit(X_train_scaled, X_train_scaled, epochs=50, batch_size=32,
              shuffle=True, validation_data=(X_test_scaled, X_test_scaled), verbose=0)

    return sdae

def evaluate_model(model, X_train_sdae, y_train, X_test_sdae, y_test):
    model.fit(X_train_sdae, y_train)
    predictions = model.predict(X_test_sdae)
    accuracy = accuracy_score(y_test, predictions)
    precision, recall, f1, _ = precision_recall_fscore_support(y_test, predictions, average='binary', pos_label='yes')
    return accuracy, precision, recall, f1

# Run the experiment for defined iterations
for iteration in range(iterations):
    sss = StratifiedShuffleSplit(n_splits=1, test_size=500, random_state=iteration)
    for train_index, test_index in sss.split(bc_tcga_normalized.iloc[:, 1:-1], bc_tcga_normalized['tumor']):
        X_train = bc_tcga_normalized.iloc[train_index]
        X_test = bc_tcga_normalized.iloc[test_index]

    # Apply Random Projection (RP)
    n_components_rp = min(dim, X_train.shape[0], X_train.shape[1])
    rp = SparseRandomProjection(n_components=n_components_rp, random_state=iteration)
    X_train_rp = rp.fit_transform(X_train.iloc[:, 1:-1])
    X_test_rp = rp.transform(X_test.iloc[:, 1:-1])

    # Apply Standard Scaler
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train_rp)
    X_test_scaled = scaler.transform(X_test_rp)

    # Train SDAE
    sdae = train_sdae(X_train_scaled, X_test_scaled)
    X_train_sdae = sdae.predict(X_train_scaled)
    X_test_sdae = sdae.predict(X_test_scaled)

    # Train and evaluate SVM
    svm = SVC(kernel='linear', C=1.0, gamma='auto')
    metrics_svm = evaluate_model(svm, X_train_sdae, X_train['tumor'], X_test_sdae, X_test['tumor'])
    avg_metrics['SVM']['accuracy'].append(metrics_svm[0])
    avg_metrics['SVM']['precision'].append(metrics_svm[1])
    avg_metrics['SVM']['recall'].append(metrics_svm[2])
    avg_metrics['SVM']['f1'].append(metrics_svm[3])

    # Train and evaluate SVM-RBF
    svm_rbf = SVC(kernel='rbf', C=1.0, gamma='auto')
    metrics_svm_rbf = evaluate_model(svm_rbf, X_train_sdae, X_train['tumor'], X_test_sdae, X_test['tumor'])
    avg_metrics['SVM-RBF']['accuracy'].append(metrics_svm_rbf[0])
    avg_metrics['SVM-RBF']['precision'].append(metrics_svm_rbf[1])
    avg_metrics['SVM-RBF']['recall'].append(metrics_svm_rbf[2])
    avg_metrics['SVM-RBF']['f1'].append(metrics_svm_rbf[3])

    # Train and evaluate ANN
    ann = MLPClassifier(hidden_layer_sizes=(100,), max_iter=1000, random_state=iteration)
    metrics_ann = evaluate_model(ann, X_train_sdae, X_train['tumor'], X_test_sdae, X_test['tumor'])
    avg_metrics['ANN']['accuracy'].append(metrics_ann[0])
    avg_metrics['ANN']['precision'].append(metrics_ann[1])
    avg_metrics['ANN']['recall'].append(metrics_ann[2])
    avg_metrics['ANN']['f1'].append(metrics_ann[3])

# Calculate and print average scores
for model in avg_metrics:
    print(f"\n{model} Results:")
    print(f"Average Accuracy: {np.mean(avg_metrics[model]['accuracy'])}")
    print(f"Average Precision: {np.mean(avg_metrics[model]['precision'])}")
    print(f"Average Recall: {np.mean(avg_metrics[model]['recall'])}")
    print(f"Average F1 Score: {np.mean(avg_metrics[model]['f1'])}")
```

```python
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, log_loss
from tensorflow.keras.layers import Dense, Dropout, GaussianNoise
from tensorflow.keras.models import Sequential
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.regularizers import l2
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Define features and target variable
X = gse25066_normalized.drop(columns=['Hybridization REF', 'tumor'])
y = gse25066_normalized['tumor']

# Normalize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Feature Selection: Apply SelectKBest to select top k features
k = 100  # You can change the number of selected features based on your needs
selector = SelectKBest(score_func=f_classif, k=k)
X_selected = selector.fit_transform(X_scaled, y)

# Stratified K-Fold Cross-Validation
kf = StratifiedKFold(n_splits=5, shuffle=True, random_state=5)

# Store metrics
accuracies = []
train_losses = []
val_losses = []

# For each fold, apply FS + SDAE and calculate losses/accuracy
for train_index, val_index in kf.split(X_selected, y):
    X_train_cv, X_val_cv = X_selected[train_index], X_selected[val_index]
    y_train_cv, y_val_cv = y.iloc[train_index], y.iloc[val_index]

    # Build SDAE Model
    sdae = Sequential()
    sdae.add(GaussianNoise(0.1, input_shape=(X_train_cv.shape[1],)))  # Add Gaussian noise to input

    sdae.add(Dense(units=200, activation='relu'))
    sdae.add(GaussianNoise(0.1))
    sdae.add(Dense(units=100, activation='relu'))
    sdae.add(GaussianNoise(0.1))
    sdae.add(Dense(units=50, activation='relu'))
    sdae.add(Dense(units=100, activation='relu'))
    sdae.add(Dense(units=200, activation='relu'))
    sdae.add(Dense(units=X_train_cv.shape[1], activation='linear'))

    sdae.compile(optimizer='adam', loss='mean_squared_error')

    # Fit the SDAE model
    sdae.fit(X_train_cv, X_train_cv, epochs=100, batch_size=32, shuffle=True, validation_data=(X_val_cv, X_val_cv), verbose=0)

    # Apply SDAE to reduce dimensionality
    X_train_sdae = sdae.predict(X_train_cv)
    X_val_sdae = sdae.predict(X_val_cv)

    # Build the ANN classification model
    model = Sequential()
    model.add(Dense(32, input_dim=X_train_sdae.shape[1], activation='relu'))
    model.add(Dropout(0.4))  # Increase dropout to reduce overfitting
    model.add(Dense(16, activation='relu', kernel_regularizer=l2(0.01)))  # L2 regularization
    model.add(Dropout(0.4))
    model.add(Dense(1, activation='sigmoid'))  # Binary classification

    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

    # Early stopping to avoid overfitting
    early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

    # Fit the ANN model
    history = model.fit(X_train_sdae, y_train_cv, epochs=100, batch_size=64,
                        validation_data=(X_val_sdae, y_val_cv), verbose=0, callbacks=[early_stopping])

    # Predict on validation set
    y_val_pred = model.predict(X_val_sdae)
    y_val_pred_classes = (y_val_pred > 0.5).astype(int)

    # Calculate accuracy
    accuracy = accuracy_score(y_val_cv, y_val_pred_classes)
    accuracies.append(accuracy)

    # Log loss
    val_loss = log_loss(y_val_cv, y_val_pred)
    val_losses.append(val_loss)

    # Train loss (using predictions on training set)
    train_probas = model.predict(X_train_sdae)
    train_loss = log_loss(y_train_cv, train_probas)
    train_losses.append(train_loss)

# Calculate average accuracy and log loss
avg_accuracy = np.mean(accuracies)
avg_train_loss = np.mean(train_losses)
avg_val_loss = np.mean(val_losses)

print(f'Average Validation Accuracy: {avg_accuracy * 100:.2f}%')
print(f'Average Training Log Loss: {avg_train_loss:.4f}')
print(f'Average Validation Log Loss: {avg_val_loss:.4f}')

# Confusion Matrix and Classification Report for the last fold
cf_matrix = confusion_matrix(y_val_cv, y_val_pred_classes)
print('\nConfusion Matrix:')
print(cf_matrix)
print('\nClassification Report:')
print(classification_report(y_val_cv, y_val_pred_classes))

# Plotting the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cf_matrix, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()

# Plotting Loss Curves
epochs = range(1, len(history.history['loss']) + 1)

plt.figure(figsize=(12, 6))
plt.plot(epochs, history.history['loss'], label='Training Log Loss', marker='o')
plt.plot(epochs, history.history['val_loss'], label='Validation Log Loss', marker='x')
plt.title('Training and Validation Log Loss')
plt.xlabel('Epochs')
plt.ylabel('Log Loss')
plt.legend()
plt.show()
```
